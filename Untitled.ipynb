```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("1. Weather Data.csv")
df.head()
```

```python
from scipy import stats

numerical_cols = ['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 
                  'Wind Speed_km/h', 'Visibility_km', 'Press_kPa']

outliers = {}

for col in numerical_cols:
    z_scores = np.abs(stats.zscore(df[col]))
    outliers[col] = (z_scores > 3).sum()

outliers
```

```python
df.columns
```

```python
plt.figure(figsize=(12,8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f"{col} Boxplot")

plt.tight_layout()
plt.show()
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

df['Year'] = df['Date/Time'].dt.year
df['Month'] = df['Date/Time'].dt.month
df['Day'] = df['Date/Time'].dt.day
df['Hour'] = df['Date/Time'].dt.hour
df['Weekday'] = df['Date/Time'].dt.day_name()
df['Temp_F'] = (df['Temp_C'] * 9/5) + 32
```

```python
df.head()
```

```python
def simplify_weather(condition):
    condition = condition.lower()
    
    if "clear" in condition:
        return "Clear"
    elif "cloud" in condition:
        return "Cloudy"
    elif "fog" in condition:
        return "Fog"
    elif "rain" in condition or "drizzle" in condition:
        return "Rain"
    elif "snow" in condition or "freezing" in condition:
        return "Snow"
    elif "thunder" in condition:
        return "Thunderstorm"
    elif "haze" in condition or "smoke" in condition:
        return "Haze/Smoke"
    else:
        return "Other"

df["Weather_Category"] = df["Weather"].apply(simplify_weather)
```

```python
df["Weather_Category"].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Weather_Category', order=df['Weather_Category'].value_counts().index)
plt.title("Distribution of Weather Categories")
plt.xlabel("Weather Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_temp = df.resample('ME', on='Date/Time')['Temp_C'].mean()

plt.figure(figsize=(12,5))
monthly_temp.plot()
plt.title("Monthly Average Temperature Trend")
plt.xlabel("Year")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
weather_month = pd.crosstab(df['Month'], df['Weather_Category'])

plt.figure(figsize=(12,6))
sns.heatmap(weather_month, annot=True, cmap="Blues")
plt.title("Weather Category Frequency by Month")
plt.ylabel("Month")
plt.xlabel("Weather Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h',
                'Visibility_km','Press_kPa']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Temp_C', y='Rel Hum_%', alpha=0.3)
plt.title("Temperature vs Relative Humidity")
plt.show()
```
